```python
import random
xox = {'T1': ' ', 'T2':' ','T3':' ',
      'M1': ' ', 'M2':' ','M3':' ',
      'D1': ' ', 'D2':' ','D3':' ',}
player1 = 1
tatal = 0
end_check=0
#دالة المقارنة للمعلرفة الفائز
def check():
    #اول احتمال افقي للفوز 
    if xox['T1'] == 'X' and xox['T2'] == 'X' and xox['T3'] == 'X':
        print("Player one Win!")
        return 1
    #تاني احتمال افقي للفوز 
    if xox['M1'] == 'X' and xox['M2'] == 'X' and xox['M3'] == 'X':
        print("Player one Win!")
        return 1
    #ثالث احتمال افقي للفوز 
    if xox['D1'] == 'X' and xox['D2'] == 'X' and xox['D3'] == 'X':
        print("Player one Win!")
        return 1
    #أول احتمال رئسي للفوز 
    if xox['T1'] == 'X' and xox['M1'] == 'X' and xox['D1'] == 'X':
        print("Player one Win!")
        return 1
    #ثاني احتمال رئسي للفوز 
    if xox['T2'] == 'X' and xox['M2'] == 'X' and xox['D2'] == 'X':
        print("Player one Win!")
        return 1
    #ثالث احتمال رئسي للفوز 
    if xox['T3'] == 'X' and xox['M3'] == 'X' and xox['D3'] == 'X':
        print("Player one Win!")
        return 1
    # القطر الرئيسي  
    if xox['T1'] == 'X' and xox['M2'] == 'X' and xox['D3'] == 'X':
        print("Player one Win!")
        return 1
    #القطر الثانوي
    if xox['T3'] == 'X' and xox['M2'] == 'X' and xox['D1'] == 'X':
        print("Player one Win!")
        return 1
    # احتمالات فوز cp
    if xox['T1'] == 'O' and xox['T2'] == 'O' and xox['T3'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['M1'] == 'O' and xox['M2'] == 'O' and xox['M3'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['D1'] == 'O' and xox['D2'] == 'O' and xox['D3'] == 'O':
        print("Player PC Win!")
        return 1
    
    if xox['T1'] == 'O' and xox['M1'] == 'O' and xox['D1'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['T2'] == 'O' and xox['M2'] == 'O' and xox['D2'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['T3'] == 'O' and xox['M3'] == 'O' and xox['D3'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['D1'] == 'O' and xox['D2'] == 'O' and xox['D3'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['T1'] == 'O' and xox['M2'] == 'O' and xox['D3'] == 'O':
        print("Player PC Win!")
        return 1
    if xox['T3'] == 'O' and xox['M2'] == 'O' and xox['D1'] == 'O':
        print("Player PC Win!")
        return 1
    return 0


while True:
    print(xox['T1'],'|',xox['T2'],'|',xox['T3'])
    print("---------")
    print(xox['M1'],'|',xox['M2'],'|',xox['M3'])
    print("---------")
    print(xox['D1'],'|',xox['D2'],'|',xox['D3'])
    end_check = check()
    if tatal == 9 or end_check == 1:
        print("The End")
        break
    while True: 
        if player1 == 1:
            p1_input = input('player one = ')
            if p1_input.upper() in xox and xox[p1_input.upper()] == ' ':
                xox[p1_input.upper()] = 'X'
                player1 = 2
                break
            else:
                print("invalid please try agin: ")
                continue
        else: 
            lis=['t1','t2','t3','m1','m2','m3','d1','d2','d3']
            random.seed()
            cp= random.choice(lis)
            c = str(cp)
            if c.upper() in xox and xox[c.upper()] == ' ':
                #    مقارنة في حال بقى للخصم مكان واحد للفوز تلقائي تقفز لمكان الذي يريد أن يلعب به الخصم
                #ولكن يوجد خلل في المقارنة مثل اذا طبق فإنه إضافة بتكون مرتين مرة في الخانة التي يريد أن يلعب بها الخصم وأخري عشوائية 
                
                if xox['D1'] and xox['D2'] == 'X':
                    xox['D3'] = 'O'
                    
                elif xox['M1'] and xox['M2'] == 'X':
                    xox['M3'] = 'O'
           
                elif xox['T1'] and xox['M1'] == 'X':
                    xox['D1'] = 'O' 
                    
                elif xox['T2'] and xox['M2'] == 'X':
                    xox['D2'] = 'O'
                    
                elif xox['T3'] and xox['M3'] == 'X':
                    xox['D3'] = 'O'
                    
                elif xox['T1'] and xox['T2'] == 'X':
                    xox['T3'] = 'O'
                    
                elif xox['T1'] and xox['M2'] == 'X':
                    xox['D3'] = 'O'
                    
                elif xox['T3'] and xox['M2'] == 'X':
                    xox['D1'] = 'O'
                    
                xox[c.upper()] = 'O'
                player1 = 1
                break
            else:
                print("invalid please try agin PC: ")
                continue
    tatal +=1
    print("-------------------")
    print()
    
    #الادخال بيكون بشكل مفاتيح القاموس مثل :T1
```
